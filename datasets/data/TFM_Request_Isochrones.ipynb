```python
import requests
import pandas as pd 
import csv
```

```python
def request_isochrone(lati, long,time, apikey):
    URL = "https://api.mapbox.com/isochrone/v1/mapbox/walking/"+str(lati)+","+str(long)+"?contours_minutes="+str(time)+"&contours_colors=6706ce&polygons=true&access_token="+apikey
    r = requests.get(url = URL)
    data = r.json()
    df = pd.json_normalize(data['features'])
    return df["geometry.coordinates"]
```

```python
vAPIKEY=''
with open('mapboxkeys.csv', mode='r', ) as csv_file:
        csv_reader = csv.DictReader(csv_file, delimiter=',')
        for row in csv_reader:
            vAPIKEY=row["APIKEY"]
```

```python
lavanderias_seccens_df = pd.read_csv('lavanderias_seccensals.csv', sep=',', decimal='.')
```

```python
lavanderias_seccens_df["isochrone_1m"] = lavanderias_seccens_df.apply(lambda row: request_isochrone(row["laundry_lat"], row["laundry_lng"], 1, vAPIKEY), axis=1)
```
