```python
from pandas_profiling import ProfileReport
import pandas as pd 
```

```python
lavanderias_seccens_df = pd.read_csv('dtjeff_seccensals.csv', sep=',', decimal='.')
```

```python
lavanderias_seccens_df.drop(['Unnamed: 0', 'Unnamed: 0.1'], axis=1, inplace=True)
```

```python
lavanderias_seccens_df.drop(['ord_creation_date_time'], axis=1, inplace=True)
```

```python
report = ProfileReport(lavanderias_seccens_df)
report.to_file("lavanderias_report.html")
```

```python
dataset_ine_df = pd.read_csv('dataset_ine_valencia.csv', sep=',', decimal='.')
```

```python
report = ProfileReport(dataset_ine_df)
report.to_file("dataset_ine_report.html")
```
